## append
**합치고자 하는 df의 column이 같을 때**  
data_name.append(name_of_the_data_to_be_appended) → in the direction of index

```python
df1 = pd.DataFrame({'class1':[95,92,98,100], 'class2':[91,93,97,99]})
```

```python
df1
```

```python
df2 = pd.DataFrame({'class1':[87,89], 'class2':[85,90]})
```

```python
df2
```

```python
# data_name.append(name_of_the_data_to_be_appended) → in the direction of index
df1.append(df2)
```

## Join
**합치고자 하는 df의 index가 같을 때**  
concatenate two dataset in the direction of column  
data_name.join(name_of_the_data_to_join)

```python
df4 = pd.DataFrame({'class3':[93,91,95,98]})
```

```python
df4
```

```python
df1
```

```python
df1.join(df4)
```

## Merge
you need to use merge when you combine two datasets **based on the value of a single column**  

how indicates the method of merging
- left : we merge the dataset based on the left data
- right
- outer
- inner

**data_name.merge(name_of_the_data_to_merge, how= , on= based_col_name)**

```python
df_A_B=pd.DataFrame({'month':['Jan', 'Feb', 'March', 'April'],
                    'Product_A':[100,150,200,130],
                    'Product_B':[90,110,140,170]})
```

```python
df_A_B
```

```python
df_C_D=pd.DataFrame({'month':['Jan', 'Feb', 'March', 'April'],
                    'Product_C':[112,141,203,134],
                    'Product_D':[90,110,140,170]})
```

```python
df_C_D
```

```python
df_A_B.merge(df_C_D, how='outer', on='month')
```

**ex)**

```python
df_left=pd.DataFrame({'key':['A','B','C'], 'left':[1,2,3]})
```

```python
df_right=pd.DataFrame({'key':['A','B','D'], 'right':[4,5,6]})
```

```python
df_left
```

```python
df_right
```

```python
df_left.merge(df_right, how = 'left', on = 'key')
```

```python
df_left.merge(df_right, how = 'right', on = 'key')
```

```python
df_left.merge(df_right, how = 'inner', on = 'key')
```

```python
df_left.merge(df_right, how = 'outer', on = 'key')
```
