###       (Write multiple lines in a new file)
Write a multiple lines using loop

create a new file

```python
f=open('multipleLines.txt', 'w') 
```

**for loop to build a multiplication table (2)**

create multi Lines

```python
for num in range(1,10):
    string = '2*{0}={1}\n'.format(num,2*num)
    f.write(string)
```

**for loop   **

```python
f = open('multipleLines.txt', 'r')
for num in range(9):
    line = f.readline()
    print(line, end='')
f.close()
```

## How to read a file and how to write a new file
### Structure
f = open('file_name', 'mode')  
line = f.read() or f.write()  
f.close()  

```python
f = open('myTextFile.txt', 'w')
f.write('This is not efficient')
f.close()
```

```python
!type myTextFile.txt
```

### with open     
with open  f.close()   

```python
with open('myTextFile_1.txt', 'w') as f:
    f.write('This is efficient')
```

```python
!type myTextFile_1.txt
```

### with open     

```python
with open('myTextFile_1.txt', 'r') as f:
    file_string = f.read()
```

```python
print(file_string)
```

### print multiple lines by using loop
ex) create multiplication table-3

```python
with open('multiplication_table3.txt', 'w') as f:
    for num in range(1,10):
        format_string = '3*{0}={1}\n'.format(num, 3*num)
        f.write(format_string)
```

```python
!type multiplication_table3.txt
```

```python
with open('multiplication_table3.txt', 'r') as f:
    for num in range(9):
        format_string = f.readline()
        print(format_string, end ='')
```
