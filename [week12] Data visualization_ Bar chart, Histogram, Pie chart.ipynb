```python
import matplotlib.pyplot as plt
```

## bar chart
before와 after를 다룰 때 유용

```python
# ID before_diet after_diet
# M_1, 80, 70
# M_2, 90, 85
# M_3, 70, 67
```

```python
member_IDs=['m_1','m_2','m_3'] # ID
before_diet=[80,90,70]
after_diet=[70,85,67]
```

plt.bar(x,height)

```python
import numpy as np
index = np.arange(3)
```

```python
plt.bar(index,before_diet)
plt.show()
```

### tick_label

```python
plt.bar(index,before_diet, tick_label=member_IDs)
plt.show()
```

### color

```python
plt.bar(index,before_diet, tick_label=member_IDs, color ='r')
plt.show()
```

```python
plt.bar(index,before_diet, tick_label=member_IDs, color ='c')
plt.show()
```

```python
colors = ['b','r','g']
plt.bar(index,before_diet, tick_label=member_IDs, color =colors, width=0.5)
plt.show()
```

### width
width in (0,1)

```python
plt.bar(index,before_diet, tick_label=member_IDs, color ='c', width=0.5)
plt.show()
```

## Compare before and after - bar chart

```python
plt.bar(index, before_diet, tick_label=member_IDs, color = 'r', width=0.5)
plt.bar(index, after_diet, tick_label=member_IDs, color = 'g', width=0.5)
plt.show()
```

```python
plt.bar(index, before_diet, tick_label=member_IDs, color = 'r', width=0.3)
plt.bar(index+0.3, after_diet, tick_label=member_IDs, color = 'g', width=0.3)
plt.show()
```

### align='edge'

```python
plt.bar(index, before_diet, tick_label=member_IDs, color = 'r', width=0.3, align='edge')
plt.bar(index+0.3, after_diet, tick_label=member_IDs, color = 'g', width=0.3, align='edge')
plt.show()
```

```python
plt.bar(index, before_diet, tick_label=member_IDs, color = 'r', width=0.3, align='edge')
plt.bar(index+0.3, after_diet, tick_label=member_IDs, color = 'g', width=0.3, align='edge')
plt.legend()
plt.title('Effect of diet')
plt.xlabel('ID')
plt.ylabel('Weight')
plt.show()
```

## pie chart
distribution of categorical variable  
plt.pie(x)

```python
fruit=['apple', 'banana', 'strawberry','orange']
preference=[7,6,3,2]
plt.pie(preference)
plt.show()
```

### labels

```python
plt.pie(preference, labels=fruit)
```

### autopct

```python
plt.pie(preference, labels=fruit, autopct='%0.1f')
# 0.1=1 decimal place(소수점 첫째자리)
# 0.2=2 decimal place(소수점 둘째자리)
plt.show()
```

```python
plt.pie(preference, labels=fruit, autopct='%0.1f%%')
# 0.1=1 decimal place(소수점 첫째자리)
# 0.2=2 decimal place(소수점 둘째자리)
plt.show()
```
