## plt.figure
**figure**란 그래프의 창, 즉 window를 뜻함  
**plt.figure()** → we create a new window

```python
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.plot(x,y1)
plt.figure()
plt.plot(x,y2)
```

```python
plt.figure(1)
plt.plot(x,y1)
plt.figure(2)
plt.plot(x,y2)
plt.figure(3)
plt.plot(x,y3)
```

```python
x=np.arange(-5,5,0.1)
```

```python
y1=x**2-2
y2=20+np.cos(x)**2
```

```python
plt.figure(1)
plt.plot(x,y1)

plt.figure(2)
plt.plot(x,y2)
```

## subplot
**plt.subplot(m,n,p)**  
- m = the number of rows
- n = the number of column
- p = the label of entry

```python
# 2-by-2 plot
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
x=np.arange(0,10,0.5)
y1=0.3*(x-5)**2+1
y2=-1.5*x+3
y3=np.sin(x)**2
y4=10*np.exp(x)+1
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
```

## fmt
**fmt = [marker] [color] [line style]**

```python
x=np.arange(0,5,1)
y1=x
y2=x+1
y3=x+2
```

```python
plt.plot(x,y1,x,y2,x,y3)
```

```python
plt.plot(x,y1,'r',x,y2,'y',x,y3,'g')
```

```python
plt.plot(x,y1,'r-',x,y2,'y:',x,y3,'g--')
```

```python
plt.plot(x,y1,'or-',x,y2,'>y:',x,y3,'sg--')
```
