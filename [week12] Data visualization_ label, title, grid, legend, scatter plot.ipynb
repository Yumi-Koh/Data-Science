```python
import numpy as np
import matplotlib.pyplot as plt
```

## label

```python
x=np.arange(-4,5,0.5)
y=2*x**3

plt.plot(x,y)
plt.show()
```

how to put the label on the x-axis and y-axis

```python
x=np.arange(-4,5,0.5)
y=2*x**3

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

## title of figure

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Title of figure')
plt.show()
```

## grid
plt.grid()

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Title of figure')
plt.grid(True)
plt.show()
```

## Scatter plot


```python
height=[165,177,160,180,185]
weight=[62,67,55,74,90]
```

```python
plt.scatter(height,weight)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & weight')
```

```python
plt.scatter(height,weight)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & weight')
plt.grid(True)
```

## plt.scatter(x,y,s=size_n, c=colors, marker='marker_string', alpha=transparency)

### size

```python
plt.scatter(height,weight,s=50)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & weight')
plt.grid(True)
```

```python
plt.scatter(height,weight,s=100)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & weight')
plt.grid(True)
```

### color

```python
plt.scatter(height,weight,s=100,c='red')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & weight')
plt.grid(True)
```

### marker

```python
plt.scatter(height,weight,s=100,c='red',marker='s')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & weight')
plt.grid(True)
```

### alpha

```python
plt.scatter(height,weight,s=100,c='red',marker='s', alpha=0.5)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & weight')
plt.grid(True)
```

### Q. plot of city based on the size of population


```python
city = ['Seoul', 'Incheon', 'Gwangju', 'Taegu', 'Pusan']
lat = [37.56, 37.45, 35.16, 35.87, 35.18]
lon = [126.97, 126.70, 126.85, 128.60, 129.70]
pop = [16154, 2751, 2995, 128.60, 4454] #people/km^2
```

```python
pop_den = np.array(pop)*0.2
```

```python
pop_den
```

```python
plt.scatter(lon, lat, s=pop_den)
```

```python
colors = ['r','g','b','c','k']
plt.scatter(lon,lat,s=pop_den,c=colors)
```

```python
colors = ['r','g','b','c','k']
plt.scatter(lon,lat,s=pop_den,c=colors)
plt.text(126.97,37.56,'Seoul')
```

```python
colors = ['r','g','b','c','k']
plt.scatter(lon,lat,s=pop_den,c=colors,alpha=0.3)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('populatio density of each city')
for x,y,name in zip(lon,lat,city):
    plt.text(x,y,name)
plt.show()
```
