## Chaning the type of data: convert the type of the ndarray data
### obj_name.astype(data_type)

```python
string = np.array(['1.5', '0.62', '2','3.14'])
```

```python
number = string.astype(float)
```

```python
number.dtype
```

```python
string2 = np.array(['1','2','3','4'])
```

```python
integer = string2.astype(int)
```

```python
integer.dtype
```
