### ▶새로운 파일에 여러 줄 쓰는 법 (Write multiple lines in a new file)
Write a multiple lines using loop

create a new file

```python
f=open('multipleLines.txt', 'w') 
```

**for loop to build a multiplication table (구구단2단)**

create multi Lines

```python
for num in range(1,10):
    string = '2*{0}={1}\n'.format(num,2*num)
    f.write(string)
```

```python
f.close()
```

show the contents of 'multipleLines.txt'

```python
!type multipleLines.txt
```

### ▶여러 줄 읽는 법 (read some of the multiple lines)
**readline()** : multipleLines.txt의 가장 **첫 번째 줄**이 화면에 출력

```python
f = open('multipleLines.txt', 'r')
```

```python
line = f.readline()
```

```python
f.close()
```

```python
print(line)
```

### What if we want to read the first 2 lines?

```python
f = open('multipleLines.txt', 'r')
line1 = f.readline()
line2 = f.readline()
f.close()
```

print(line1) 은 \n을 포함한 상태

```python
print(line1)
print(line2)
```

\n을 포함하지 않게 하기 위해서

```python
print(line1, end = '')
print(line2, end = '')
```

**for loop를 사용하여 여러줄 읽기**

```python
f = open('multipleLines.txt', 'r')
for num in range(9):
    line = f.readline()
    print(line, end='')
f.close()
```

## How to read a file and how to write a new file
### Structure
f = open('file_name', 'mode')  
line = f.read() or f.write()  
f.close()  

```python
f = open('myTextFile.txt', 'w')
f.write('This is not efficient')
f.close()
```

```python
!type myTextFile.txt
```

### ▶with open을 사용하여 새로운 파일 쓰는 법
with open을 이용하면 f.close()를 쓰지 않아도 됨

```python
with open('myTextFile_1.txt', 'w') as f:
    f.write('This is efficient')
```

```python
!type myTextFile_1.txt
```

### ▶with open을 사용하여 이전 파일 읽는 법

```python
with open('myTextFile_1.txt', 'r') as f:
    file_string = f.read()
```

```python
print(file_string)
```

### print multiple lines by using loop
ex) create multiplication table-3단

```python
with open('multiplication_table3.txt', 'w') as f:
    for num in range(1,10):
        format_string = '3*{0}={1}\n'.format(num, 3*num)
        f.write(format_string)
```

```python
!type multiplication_table3.txt
```

```python
with open('multiplication_table3.txt', 'r') as f:
    for num in range(9):
        format_string = f.readline()
        print(format_string, end ='')
```
