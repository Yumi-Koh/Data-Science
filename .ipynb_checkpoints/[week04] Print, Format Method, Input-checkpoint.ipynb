## Format method

### ▶% using method
print('%type' % data)
- s(string)
- d(integer)
- f(float)

```python
data = 'Byunghoon'
print('%s' % data)
```

```python
score=100
print('My score is %d' % score +'.')
```

```python
data1=3
data2=3.141592
print('Radius: %d, Circumference: %f' % (data1, data2))
```

### ▶string.format()

```python
animal_0='cat'
animal_1='dog'
animal_2='fox'
print('Animal: {0}'.format(animal_0))
```

```python
animal_0='cat'
animal_1='dog'
animal_2='fox'
print('Animal: {0} {1} {2}'.format(animal_0, animal_1, animal_2))
```

```python
print('Animal: {0} {2} {1}'.format(animal_0, animal_1, animal_2))
```
