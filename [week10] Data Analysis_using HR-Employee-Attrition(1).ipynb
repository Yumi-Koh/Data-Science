### 2. Select a row

### 3. Select a index
we know the name of index  
**data_name.loc[index_name]**  
**data_name.loc[starting_index_name:ending_index_name]**

```python
#index값을 직접 명시
data.loc[[2,5]]
```

```python
#index의 범위
data.loc[2:5]
```

### 4. We can specify the rows and columns simultaneously
1. data_name[columns_name][index_location]
2. data_name.loc[index_name][column_name]
3. data_name.iloc[index_loc, column_loc]

```python
# 1) data_name[columns_name][index_location]
data['BusinessTravel'][5:7]
```

```python
# 2) data_name.loc[index_name][column_name]
data.loc[5:7]['BusinessTravel']
```

```python
# 3) data_name.iloc[index_loc, column_loc]
# You just specify the column_loc and index_loc
data.iloc[:20,:10]
```
